```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Files/cluster_mpg.csv")
```

```python
df
```

```python
df.describe()
```

```python
df_w_dummies = pd.get_dummies(df.drop("name", axis=1))
```

```python
df_w_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_w_dummies)
```

```python
scaled_data
```

```python
scaled_df=pd.DataFrame(scaled_data,columns=df_w_dummies.columns)
```

```python
scaled_df
```

```python
sb.heatmap(scaled_df)
```

```python
sb.clustermap(scaled_df,row_cluster=False)
```

```python
sb.clustermap(scaled_df,col_cluster=False)
```
