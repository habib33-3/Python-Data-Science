```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("../../Files/mouse_viral_study.csv")
```

```python
from sklearn.svm import SVC
```

```python
X = df.drop("Virus Present", axis=1)
```

```python
y = df["Virus Present"]
```

```python
model= SVC(kernel="rbf",C=1)
```

```python
model.fit(X,y)
```
